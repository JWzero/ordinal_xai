```python
from ordinal_xai.models.obd import OBD
from ordinal_xai.interpretation.ice import ICE
from ordinal_xai.interpretation.ice_prob import ICEProb

from ordinal_xai.utils.data_utils import load_data
import effector
import numpy as np
import pandas as pd


X,y = load_data("ordinal_xai/data/winequality-red.csv")
model = OBD()
model.fit(X,y)


ice = ICE(model, X, y)
ice.explain(plot=True,feature_subset=[3])
'''
ice_prob = ICEProb(model, X, y) 
ice_prob.explain(plot=True,feature_subset=[6])
'''
def predict(x): #wrapper function
    return np.array(model.predict(pd.DataFrame(x))).squeeze()   


#use effector
r_pdp = effector.RegionalPDP(np.array(X),predict,feature_names=list(X.columns))


r_pdp.summary(features=3)

r_pdp.plot(feature=3,node_idx=1)
r_pdp.plot(feature=3,node_idx=2)




```
